```python
#See how many unique values each categorial variable has 
cat_feat = accepted_loans.select_dtypes('object').columns.values
accepted_loans[cat_feat].nunique().sort_values()
```

```python
#As we can see, title, desc and emp_title have a lot of different values.Would cause memory problems.
#Would be value in NLP with these columns, but beyond scope for us
#Also we want to get rid of sub_grade (we know is highly correlated with grade)
accepted_loans = accepted_loans.drop(['title', 'desc','emp_title','sub_grade'], axis=1, errors='ignore')
```

```python
#Used to calculate pearson correlation between the various column
import numpy as np
from itertools import combinations
from scipy.stats import pearsonr
num_feat = accepted_loans.select_dtypes('number').columns.values
comb_num_feat = np.array(list(combinations(num_feat, 2)))
corr_num_feat = np.array([])
for comb in comb_num_feat:
    corr = pearsonr(accepted_loans[comb[0]], accepted_loans[comb[1]])[0]
    corr_num_feat = np.append(corr_num_feat, corr)
```

```python
#Viewing these
high_corr_num = comb_num_feat[np.abs(corr_num_feat) >= 0.9]
high_corr_num
```

```python
#Dropping the second one in each pair bc I wanted to keep loan_amnt 
accepted_loans = accepted_loans.drop(np.unique(high_corr_num[:, 1]), axis=1, errors='ignore')
```

```python
#How many loans are in each status
my_list = accepted_loans['loan_status'].value_counts()
my_list
```

```python
#We considered loans with status Current, Fully Paid, and In Grace Period as good. 
#All the other statuses are considered as bad loans (late 16-30 days, Late 31-120 days, or default) 
y = accepted_loans['loan_status'].copy()
y = y.isin(['Current', 'Fully Paid','In Grace Period','Does not meet the credit policy. Status:Fully Paid']).astype('int')
y.value_counts()
```

```python
#Only concerned with high risk/high reward loans. So we focus on grades with E F and G
#Remove grade/interest rate after selecting the high risk ones (and therefore high interest rate loans)
#Also remove loan status as that's now our y variable
X = accepted_loans[(accepted_loans.grade == 'E') | (accepted_loans.grade == 'F') | (accepted_loans.grade == 'G')].copy()
X.drop
X = X.drop(['grade', 'int_rate','loan_status','addr_state'], axis=1, errors='ignore')
```

```python
#Also drop ID since it is no longer relevant
X = X.drop(['id'], axis = 1)
```

```python
X.head()
```

```python
#Only get the y values for which there is an X
y_mod = y[X.index]
```

```python
#Combining the y and X variables into one dataframe
X_2 = pd.concat([y_mod, X], axis=1)
```

```python
#normalizing the data
from sklearn import preprocessing
X_3 = X_2.drop(['application_type', 'home_ownership','purpose'],axis=1)
X_array = X_3.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(X_array)
X_norm=pd.DataFrame(x_scaled, columns=X_3.columns)
X_norm.head()
```
